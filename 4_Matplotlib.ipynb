```python
# plots the scores of the two students in five different subjects. subjects as x-axis and marks as y-axis
s1=[56,78,92,95,78]
s2=[77,89,91,23,45]
subject=["MATHS","SCIENCE","ENGLISH","TELUGU","SOCIAL"]
plt.plot(subject,s1,label="s1")
plt.plot(subject,s2,label="s2")
plt.title("Marks of subjects")
plt.legend(loc="best")
```

```python
s1=[56,78,92,95,78]
s2=[77,89,91,23,45]
subject=["MATHS","SCIENCE","ENGLISH","TELUGU","SOCIAL"]
plt.subplot(2,1,1)
plt.plot(subject,s1,label="s1",color="c")
plt.legend()
plt.subplot(2,1,2) #row ,column,spliting position in graph paper
plt.plot(subject,s2,label="s2",color="b")
plt.legend()
```

**Problem : 3**

```python
import numpy as np
import matplotlib.pyplot as plt
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("Revenues of company A",A)
print("Revenues of company B",B)
proofA=np.diff(A)
print("Profits of A",proofA)
proofB=np.diff(B)
print("Profits of B",proofB)
year=["2019-20","2020-21","2021-22","2022-23"]
plt.subplot(2,1,1)
plt.bar(year,proofA,label="Profit-A",color="k")
plt.legend(loc="best")
plt.subplot(2,1,2)
plt.bar(year,proofB,label="Profit-B",color="c")
plt.legend(loc="best")
```
