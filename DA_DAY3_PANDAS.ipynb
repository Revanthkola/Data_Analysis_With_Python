```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(data=tips,y="tip",x="total_bill",hue="smoker",palette="deep")
#hue - creating difference based on a column via color
```

```python
sns.relplot(data=tips,y="tip",x="total_bill",hue="time",palette="pastel")
```

```python
sns.relplot(data=tips,y="tip",x="total_bill",hue="day",palette="colorblind",kind="line")
```

```python
sns.relplot(data=tips,y="tip",x="total_bill",hue="size",palette="dark")
```

```python
sns.relplot(data=tips,y="tip",x="total_bill",hue="sex",palette="muted")
```

```python
sns.relplot(data=tips,y="tip",x="total_bill",hue="time",style="day",palette="bright")
```

```python
sns.catplot(data=tips,x="total_bill",y="day")
```

```python
sns.catplot(data=tips,x="time",y="total_bill",kind="violin")
```

```python
sns.catplot(data=tips,x="time",y="total_bill",kind="violin",hue="smoker")
```
